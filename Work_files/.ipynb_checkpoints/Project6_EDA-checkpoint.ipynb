```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Shinemet\Ironhack\Projects\Project6\Work_files\Payments_clean.csv", sep=';')
```

```python
data
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in data:
    if data[col].dtype=="object":
        print(col)
        data[col] = encoder.fit_transform(data[col])

data
```

```python
data.drop(columns=['Debtor', 'Type', 'Reference'], inplace=True)
```

```python
data.describe()
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(data.corr(), annot=True, square=True)

plt.show()
```

```python
country = data.groupby('BIC_Country').mean()
country.sort_values(by='Duration_days', ascending=False, inplace=True)
country
```
